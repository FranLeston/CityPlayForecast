Average of sales by day

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#Set the seaborn theme
sns.set_style("whitegrid")
```

```python
df_sales = pd.read_csv('data/db_load_files/clean_data.csv')
df_sales
```

```python
fig_dims = (20, 4)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x="year", y="total_sales", hue="day_of_week",ax=ax, data=df_sales)
```

```python
cat_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df2 = df_sales.groupby('day_of_week').mean().reindex(cat_days) 
df2.reset_index(inplace=True)
sns.barplot(x='day_of_week', y='total_sales', data=df2);
```

```python
fig_dims = (20, 4)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x="year", y="total_sales", hue="day_type",ax=ax, data=df_sales)
```

```python
fig_dims = (20, 4)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x="year", y="total_sales", hue="month_name",ax=ax, data=df_sales)
```

```python
fig_dims = (20, 4)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x="year", y="total_sales", hue="did_rain",ax=ax, data=df_sales)
```
