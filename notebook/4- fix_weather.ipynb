<h2>Weather Dataset Cleaning</h2>

```python
import pandas as pd
import numpy as np

import datetime
import seaborn as sns
from matplotlib import rcParams
import matplotlib.pyplot as plt
# figure size in inches
rcParams['figure.figsize'] = 30,8
```

```python
df_climate = pd.read_excel ('../data/raw_data/raw_climate.xlsx')
df_climate.head()
```

<h2>There is lots of data here..lets give the columns meaningful names</h2>

```python
df_climate.rename(columns={'Y': 'year','M': 'month', 'D': 'day', 'T':'average_temp', 'PP':'total_precip_mm', 'RA':'did_rain', 'SN': 'did_snow' }, inplace=True)
df_climate
```

<h2>Make a date time from these 3 columns:</h2>

```python
df_climate['date'] = pd.to_datetime(df_climate[['year', 'month', 'day']])
```

```python
df_climate
```

<h2>Only get date range I have sales data for...</h2>

```python
mask = (df_climate['date'] > '2015-09-20') & (df_climate['date'] <= '2021-05-11')
```

```python
new_df_climate = df_climate.loc[mask]
```

```python
#List all my missing dates
pd.date_range(start = '2015-09-21', end = '2021-05-11' ).difference(new_df_climate.date)
```

```python
new_df_climate.tail(25)
```

<h2>Lets get the columns we are interested in..Temperature, Rainfall, Snow and Rain amount</h2>

```python
new_df_climate = new_df_climate[["date", "average_temp", "total_precip_mm", "did_rain", "did_snow"]].set_index("date")
```

```python
new_df_climate.sample(100)
```

```python
new_df_climate.dtypes
```

<h2>If there is rainfall amount on certain day, that means it did in fact rain..change did_rain in thos cases to 1's.</h2>

```python
new_df_climate.loc[new_df_climate.total_precip_mm > 0, 'did_rain'] = 1
```

<h2>Lets save this!</h2>

```python
new_df_climate.to_csv("../data/raw_data/climate_clean.csv" , index=True)
```
