```python
data_frames = [bolera_trend, bowling_trend, minigolf_trend]
```

```python
trends_merged= reduce(lambda left,right: pd.merge(left,right,on='Month'), data_frames)
```

```python
trends_merged.rename(columns={'bolera: (Community of Madrid)': 'bolera','bowling: (Community of Madrid)' :'bowling' , 'minigolf: (Community of Madrid)':'minigolf' }, inplace=True)
```

```python
def getTrends(x, trend):
    date_search = x['date'][0:7]
    row = trends_merged[trends_merged['Month'] == date_search]
    return row.iloc[0][trend]

trends_list = ["bolera", "bowling", "minigolf"]

for trend in trends_list:
    df2[trend + '_trends'] = df2.apply(lambda x: getTrends(x, trend), axis=1)
```

```python
df2['is_curfew'] = df2['date'].apply(lambda x: 1 if x >= '2020-03-13' and x <= '2021-05-09' else 0)
```

```python
X = df4.drop(['total_sales'], axis=1)
y = df4['total_sales']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
models = { "ridge": Ridge(),
          "lasso": Lasso(),
          "sgd": SGDRegressor(),
          "knn": KNeighborsRegressor(),
          "gradient": GradientBoostingRegressor()
}
```

```python
for name, model in models.items():
    print(f"Entrenando modelo ---> {name}")
    model.fit(X_train,y_train)
    print(f"He acabado :)")
```

```python
for name, model in models.items():
    y_pred = model.predict(X_test)
    print(f"--------{name}--------")
    print("MAE: ", mean_absolute_error(y_test, y_pred))
    print("MSE: ", mean_squared_error(y_test,y_pred))
    print("RMSE: ", np.sqrt(mean_squared_error(y_test,y_pred)))
    print("R2: ", r2_score(y_test,y_pred))
    print("\n")
```

```python
model = RandomForestRegressor()

params = {'n_estimators': [450],
          'max_features': ["sqrt"],
          'max_depth': [30],
          'min_samples_leaf': [2]}

grid_search = GridSearchCV(model, param_grid=params, verbose=1, n_jobs=-1,cv=5)

grid_search.fit(X_train, y_train)
```

```python
bestscore = grid_search.best_score_
print("Best GridSearch Score: ", bestscore)
best_rf = grid_search.best_estimator_
print("Best Estimator: ", best_rf)
print("Best RF SCORE: ", best_rf.score(X, y))
```

```python
X["predicted_sales"] = best_rf.predict(X)
```
